# Business Problem 1

```python
# import relevant libraries for analysis

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

# Q.1 Cleaning the data 

```python
# import Amazine_sales.csv
# only showing relevant columns 
sales_data = pd.read_csv('Amazine_sales.csv',encoding='latin1')

sales_data.head(4)
```

```python
# i think the numbers in the data frame should be to 2 decimal places for easier reading

pd.options.display.float_format ='{:.2f}'.format
sales_data.head(4)
```

```python
# shape of the data

sales_data.shape
```

```python
# the data type of each coloumn

sales_data.dtypes
```

```python
# looking for any column with no data

sales_data.isnull().sum()
```

```python
# replacing any missing data with a 0 
# there isn't any just to make sure

filled_sales_data = sales_data.fillna(0)
filled_sales_data.head(4)
```

# Data is now cleaned and ready for interpretation

# Q.2 List the top 3 sales people for each region

```python
# this is done by doing 2 groupbys on the data
# this will show the top 3 salespeople for each reason using nlargest 
# This effectively groups the data by Region and Salesperson, and counts how many times each Salesperson appears in each Region.

top_salespeople = filled_sales_data.groupby('Region')['Salesperson'].value_counts().groupby(level=0).nlargest(3)
top_salespeople
```

# Q.3 Create a calculated column called ‘Price’ 

```python
# new column called ‘Price’ (=Sales / Quantity) for each product.

filled_sales_data['Price'] = filled_sales_data['Sales'] / sales_data['Quantity']
selected_columns = ['Order ID', 'Order Date', 'Ship Date', 'Ship Mode', 'Salesperson', 'City', 'Region', 'Category', 'Sub-Category', 'Product Name', 'Sales', 'Quantity', 'Discount', 'Profit','Price']

filled_sales_data[selected_columns].head(4)
```

# Q4. the sum of ‘Profit’ for each Region using the agg method from high to low

```python
# the sum of Profit for each Region using the agg method from high to low
# I grouped by region and showed the sum of profit for each region and used sort_values to show the highest first 
# agg is used to aggregate and transform grouped data

profit_by_region = filled_sales_data.groupby('Region')['Profit'].agg('sum').sort_values(ascending=False)
profit_by_region
```

# Q5. Using groupby, display the top 5 Sales for each Category and Subcategory.

```python
# Using groupby, display the top 5 Sales for each Category and Subcategory.

top_sales_by_category = filled_sales_data.groupby(['Category', 'Sub-Category'])['Sales'].agg('sum').groupby(level=0).nlargest(5)
top_sales_by_category
```

# Q6. Using a pivot table, show the sum and mean of ‘Sales’ in terms of Segment and Category. 

```python
# created a pivot table to show the sum and mean of Sales in terms of Segment and Category. 

sales_pivot = filled_sales_data.pivot_table(index=['Segment', 'Category'], values='Sales', aggfunc={'Sales': ['sum', 'mean']})
sales_pivot
```

```python
# changing the date format using to_datetime so i can filter the data better

filled_sales_data[['Order Date', 'Ship Date']] = filled_sales_data[['Order Date', 'Ship Date']].apply(pd.to_datetime, format='%d/%m/%Y')
filled_sales_data[selected_columns].head(4)
```

# Q7. Find the Office Supplies sold in Fremont worth more than 20 dollars and sold by Ken Black in 2016.

```python
# Find the Office Supplies sold in Fremont worth more than $20 and sold by Ken Black in 2016.

office_supplies_fremont = filled_sales_data[(filled_sales_data['City'] == 'Fremont') 
                                            & (filled_sales_data['Category'] == 'Office Supplies') 
                                            & (filled_sales_data['Salesperson'] == 'Ken Black') 
                                            & (filled_sales_data['Order Date'].dt.year == 2016) 
                                            & (filled_sales_data['Sales'] > 20)]
office_supplies_fremont[selected_columns]
```

```python
# for debugging i forgot to add dt year so i removed the ship date condition to find the problem

phone_first_class_sales = filled_sales_data[(filled_sales_data['Sub-Category'] == 'Phones') 
                                            & (filled_sales_data['Ship Mode'] == 'First Class')]

phone_first_class_sales[selected_columns].head(4)
```

# Q8. Find all the salespeople who sold Phones with a discount that were shipped ‘First class’ in 2016.

```python
# Find all the salespeople who sold Phones with a discount that were shipped First class in 2016.

phone_sales_2016 = filled_sales_data[(filled_sales_data['Sub-Category'] == 'Phones') 
                                     & (filled_sales_data['Discount'] > 0) 
                                     & (filled_sales_data['Ship Mode'] == 'First Class') 
                                     & (filled_sales_data['Ship Date'].dt.year == 2016)]

phone_sales_2016[selected_columns].head(4)
```

 # Q9. Any sales visualisation (graph) that you think would help the Amazine management to interpret their sales data better. 

```python
# getting the Products that make the company lose the most money by using nsmallest
top_loss_products = filled_sales_data.nsmallest(10, 'Profit')
top_loss_products[['Profit','Product Name']]
```

```python
# Group data by Salesperson and aggregate sum Sales and Profit for each group.
# Then reset the index to make Salesperson a regular column in the resulting DataFrame.
salesperson_summary = filled_sales_data.groupby('Salesperson').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

# Find the salesperson with the highest profit by comparing each salesperson's profit to the maximum profit in the summary.
most_profit_salesperson = salesperson_summary[salesperson_summary['Profit'] == salesperson_summary['Profit'].max()]

most_profit_salesperson
```

```python
# Find the salesperson with the highest sales by comparing each salesperson's sales to the maximum sales in the summary.

most_sales_salesperson = salesperson_summary[salesperson_summary['Sales'] == salesperson_summary['Sales'].max()]

most_sales_salesperson
```
